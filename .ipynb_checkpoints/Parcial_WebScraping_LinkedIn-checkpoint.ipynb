### Importacion de librerias

````python
#codigo de inicio
import requests
from bs4 import BeautifulSoup
import pandas as pd
````

---
## Punto 1 - Inspeccion del codigo HTML de LinkedIn Jobs

Se accedio a [LinkedIn Jobs](https://www.linkedin.com/jobs/search/?keywords=data+scientist&location=United+States) en modo incognito realizando una busqueda de **'data scientist'** en **'United States (EEUU)'**.

Al inspeccionar el codigo fuente de la pagina de resultados, se identifico que cada anuncio de empleo esta contenido en un elemento `<li>` dentro de una lista `<ul>`. Cada tarjeta es un `<div class="base-card">` con los siguientes sub-elementos:

| Campo | Selector |
|---|---|
| Titulo | `h3.base-search-card__title` |
| Empresa | `h4.base-search-card__subtitle` |
| Localizacion | `span.job-search-card__location` |
| URL | `a.base-card__full-link` atributo `href` |

**El elemento que permite llegar exactamente a la lista de anuncios es:**
```html
<ul class="jobs-search__results-list">
```
Cada `<li>` dentro de esta lista corresponde a un anuncio de empleo individual.


---
## Punto 2 - Extraccion de la lista de anuncios de trabajo


````python
# URL generada por LinkedIn al buscar 'data scientist' en EEUU (endpoint publico sin login)
url = "https://www.linkedin.com/jobs/search/?keywords=data+scientist&location=United+States"

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/122.0.0.0 Safari/537.36"
    )
}

# Peticion HTTP al servidor de LinkedIn
respuesta = requests.get(url, headers=headers, timeout=20)
print(f"Codigo de respuesta HTTP: {respuesta.status_code}")

# Parseo del HTML con BeautifulSoup
soup = BeautifulSoup(respuesta.text, "html.parser")

# El elemento clave es <ul class="jobs-search__results-list">
# cada <li> dentro es un anuncio de empleo
lista_empleos = soup.find_all("li")
print(f"Elementos <li> encontrados: {len(lista_empleos)}")
````

---
## Punto 3 - Extraccion del primer anuncio

Se selecciona el primer anuncio de la lista y se extraen: **titulo**, **empresa**, **localizacion** y **URL**.


````python
# Seleccionamos el primer anuncio de la lista
primer_anuncio = lista_empleos[0]

# Titulo del trabajo
titulo = primer_anuncio.find("h3", class_="base-search-card__title")
titulo = titulo.get_text(strip=True) if titulo else "No disponible"

# Nombre de la compania
empresa = primer_anuncio.find("h4", class_="base-search-card__subtitle")
empresa = empresa.get_text(strip=True) if empresa else "No disponible"

# Localizacion (ciudad / municipio)
localizacion = primer_anuncio.find("span", class_="job-search-card__location")
localizacion = localizacion.get_text(strip=True) if localizacion else "No disponible"

# URL del anuncio
url_anuncio = primer_anuncio.find("a", class_="base-card__full-link")
url_anuncio = url_anuncio.get("href", "No disponible") if url_anuncio else "No disponible"

print("=== Primer anuncio extraido ===")
print(f"Titulo       : {titulo}")
print(f"Empresa      : {empresa}")
print(f"Localizacion : {localizacion}")
print(f"URL          : {url_anuncio}")
````

---
## Punto 4 - Extraccion de todos los anuncios al DataFrame


````python
#complete este codigo

# Lista para almacenar todos los empleos extraidos
datos = []

for anuncio in lista_empleos:
    # Titulo del trabajo
    titulo = anuncio.find("h3", class_="base-search-card__title")
    titulo = titulo.get_text(strip=True) if titulo else None

    # Nombre de la compania
    empresa = anuncio.find("h4", class_="base-search-card__subtitle")
    empresa = empresa.get_text(strip=True) if empresa else None

    # Localizacion
    localizacion = anuncio.find("span", class_="job-search-card__location")
    localizacion = localizacion.get_text(strip=True) if localizacion else None

    # URL del anuncio
    url_anuncio = anuncio.find("a", class_="base-card__full-link")
    url_anuncio = url_anuncio.get("href") if url_anuncio else None

    # Solo guardamos registros que tengan al menos titulo y empresa
    if titulo and empresa:
        datos.append({
            "titulo"      : titulo,
            "empresa"     : empresa,
            "localizacion": localizacion,
            "url"         : url_anuncio,
        })

# Creamos el DataFrame
df = pd.DataFrame(datos)

print(f"Anuncios extraidos: {len(df)}")
df.head()
````

---
## Punto 5 - Exportacion a archivo CSV


````python
# Exportamos a CSV con codificacion UTF-8 (compatible con Excel)
df.to_csv("empleos_data_scientist_eeuu.csv", index=False, encoding="utf-8-sig")

print("Archivo exportado: empleos_data_scientist_eeuu.csv")
print(f"Registros guardados: {len(df)}")
````

## Punto 6 - Cuantas ofertas contiene el dataframe y cuantos resultados hay en LinkedIn?

LinkedIn muestra **~25 anuncios por pagina** en su endpoint publico (sin login). Con una sola peticion, el DataFrame contiene como maximo **25 registros**, mientras que LinkedIn puede mostrar cientos o miles de resultados totales para la busqueda.

### Diferencia entre lo extraido y el total disponible

| | Descripcion |
|---|---|
| Registros en el DataFrame | Los que devuelve la primera pagina (~25) |
| Resultados totales en LinkedIn | Indicado en la barra superior de la busqueda |
| Anuncios por pagina | 25 |

La diferencia se debe a que **solo se hizo una peticion a la primera pagina**.

### Como extraer todos los resultados

Al revisar las URLs de cada pagina de resultados, el parametro `start` controla la paginacion:

```
Pagina 1: ...&start=0
Pagina 2: ...&start=25
Pagina 3: ...&start=50
Pagina N: ...&start=(N-1)*25
```

Para obtener **todos** los resultados se debe iterar sobre cada pagina incrementando `start` en 25, concatenando los DataFrames de cada pagina, como se ilustra a continuacion:


````python
import time

BASE_URL = (
    "https://www.linkedin.com/jobs/search/"
    "?keywords=data+scientist&location=United+States&start={start}"
)

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/122.0.0.0 Safari/537.36"
    )
}

def extraer_pagina(start):
    """Extrae los anuncios de una pagina dado su offset."""
    url = BASE_URL.format(start=start)
    respuesta = requests.get(url, headers=headers, timeout=20)
    soup = BeautifulSoup(respuesta.text, "html.parser")
    anuncios = soup.find_all("li")
    resultados = []
    for anuncio in anuncios:
        titulo       = anuncio.find("h3", class_="base-search-card__title")
        empresa      = anuncio.find("h4", class_="base-search-card__subtitle")
        localizacion = anuncio.find("span", class_="job-search-card__location")
        url_anuncio  = anuncio.find("a", class_="base-card__full-link")
        if titulo and empresa:
            resultados.append({
                "titulo"      : titulo.get_text(strip=True),
                "empresa"     : empresa.get_text(strip=True),
                "localizacion": localizacion.get_text(strip=True) if localizacion else None,
                "url"         : url_anuncio.get("href") if url_anuncio else None,
            })
    return resultados

# --- Para extraer todas las paginas, descomenta el bloque siguiente ---
# num_paginas = 10  # ajustar segun el total de resultados de la busqueda
# todos = []
# for pagina in range(num_paginas):
#     start = pagina * 25
#     print(f"Extrayendo pagina {pagina + 1} (start={start})...")
#     todos.extend(extraer_pagina(start))
#     time.sleep(3)  # pausa para respetar el servidor
#
# df_completo = pd.DataFrame(todos)
# df_completo.to_csv("empleos_data_scientist_eeuu_completo.csv", index=False, encoding="utf-8-sig")
# print(f"Total de registros extraidos: {len(df_completo)}")

print("Funcion extraer_pagina() definida.")
print("Para extraer todas las paginas, descomenta el bloque de paginacion.")
````
